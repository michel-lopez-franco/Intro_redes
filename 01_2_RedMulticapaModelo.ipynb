```python
from graphviz import Digraph

def create_neural_network():
    # Crear un nuevo grafo
    dot = Digraph(comment='Red Neuronal Multicapa')
    dot.attr(rankdir='LR')  # Dirección izquierda a derecha
    dot.attr(splines='line')  # Líneas rectas
    
    # Crear subgrafo para la capa de entrada
    with dot.subgraph(name='cluster_input') as c:
        c.attr(label='Input Layer', color='lightblue')
        # Crear nodos de entrada
        for i in range(3):
            c.node(f'i{i+1}', f'x₁', shape='circle', style='filled', fillcolor='lightblue')
        c.attr(rank='same')
    
    # Crear subgrafo para la primera capa oculta
    with dot.subgraph(name='cluster_hidden1') as c:
        c.attr(label='Hidden Layer 1', color='lightgreen')
        # Crear nodos de la primera capa oculta
        for i in range(3):
            c.node(f'h1{i+1}', f'h₁{i+1}', shape='circle', style='filled', fillcolor='lightgreen')
        c.attr(rank='same')
    
    # Crear subgrafo para la segunda capa oculta
    with dot.subgraph(name='cluster_hidden2') as c:
        c.attr(label='Hidden Layer 2', color='lightgreen')
        # Crear nodos de la segunda capa oculta
        for i in range(2):
            c.node(f'h2{i+1}', f'h₂{i+1}', shape='circle', style='filled', fillcolor='lightgreen')
        c.attr(rank='same')
    
    # Crear subgrafo para la capa de salida
    with dot.subgraph(name='cluster_output') as c:
        c.attr(label='Output Layer', color='pink')
        c.node('o1', 'y₁', shape='circle', style='filled', fillcolor='pink')
        c.attr(rank='same')
    
    # Crear conexiones entre capas
    # Conexiones de entrada a primera capa oculta
    for i in range(3):  # desde cada nodo de entrada
        for j in range(3):  # a cada nodo de la primera capa oculta
            dot.edge(f'i{i+1}', f'h1{j+1}')
    
    # Conexiones de primera capa oculta a segunda capa oculta
    for i in range(3):  # desde cada nodo de la primera capa oculta
        for j in range(2):  # a cada nodo de la segunda capa oculta
            dot.edge(f'h1{i+1}', f'h2{j+1}')
    
    # Conexiones de segunda capa oculta a salida
    for i in range(2):  # desde cada nodo de la segunda capa oculta
        dot.edge(f'h2{i+1}', 'o1')
    
    return dot

# Crear y guardar el grafo
nn = create_neural_network()

# Renderizar el grafo (genera archivo PNG)
nn.render('red_neuronal', format='png', cleanup=True)

# Para visualizar en Jupyter Notebook, puedes usar:
# nn.view()
```

```python
# Importar bibliotecas necesarias
from graphviz import Digraph
from IPython.display import display, Image

def create_neural_network():
    # Crear un nuevo grafo
    dot = Digraph(comment='Red Neuronal Multicapa')
    dot.attr(rankdir='LR')  # Dirección izquierda a derecha
    dot.attr(splines='line')  # Líneas rectas
    
    # Crear subgrafo para la capa de entrada
    with dot.subgraph(name='cluster_input') as c:
        c.attr(label='Input Layer', color='lightblue')
        for i in range(3):
            c.node(f'i{i+1}', f'x{i+1}', shape='circle', style='filled', fillcolor='lightblue')
        c.attr(rank='same')
    
    # Crear subgrafo para la primera capa oculta
    with dot.subgraph(name='cluster_hidden1') as c:
        c.attr(label='Hidden Layer 1', color='lightgreen')
        for i in range(3):
            c.node(f'h1{i+1}', f'h₁{i+1}', shape='circle', style='filled', fillcolor='lightgreen')
        c.attr(rank='same')
    
    # Crear subgrafo para la segunda capa oculta
    with dot.subgraph(name='cluster_hidden2') as c:
        c.attr(label='Hidden Layer 2', color='lightgreen')
        for i in range(2):
            c.node(f'h2{i+1}', f'h₂{i+1}', shape='circle', style='filled', fillcolor='lightgreen')
        c.attr(rank='same')
    
    # Crear subgrafo para la capa de salida
    with dot.subgraph(name='cluster_output') as c:
        c.attr(label='Output Layer', color='pink')
        c.node('o1', 'y₁', shape='circle', style='filled', fillcolor='pink')
        c.attr(rank='same')
    
    # Crear conexiones entre capas
    # Entrada a primera capa oculta
    for i in range(3):
        for j in range(3):
            dot.edge(f'i{i+1}', f'h1{j+1}')
    
    # Primera capa oculta a segunda
    for i in range(3):
        for j in range(2):
            dot.edge(f'h1{i+1}', f'h2{j+1}')
    
    # Segunda capa oculta a salida
    for i in range(2):
        dot.edge(f'h2{i+1}', 'o1')
    
    return dot

# Crear el grafo
nn = create_neural_network()

# Método 1: Visualización directa en el notebook
# Esta es la forma más simple
nn
```
