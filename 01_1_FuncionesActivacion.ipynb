
## Funciones de Activación y sus Derivadas

En esta sección, exploraremos tres funciones de activación comunes utilizadas en redes neuronales: Sigmoid, Tanh y ReLU, junto con sus derivadas.

### Funciones de Activación

1. **Sigmoid**: La función Sigmoid es una función no lineal que mapea cualquier valor real a un rango entre 0 y 1. Es útil para modelos donde necesitamos predecir la probabilidad como una salida.
    
    $$\text{sigmoid}(x) = \frac{1}{1 + e^{-x}}$$
    
    
2. **Tanh**: La función Tanh es otra función no lineal que mapea cualquier valor real a un rango entre -1 y 1. Es una versión escalada de la función Sigmoid.

    $$
    \text{tanh}(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
    $$

3. **ReLU (Rectified Linear Unit)**: La función ReLU es una función no lineal que mapea cualquier valor negativo a 0 y cualquier valor positivo a sí mismo. Es ampliamente utilizada en redes neuronales profundas debido a su simplicidad y eficiencia computacional.

    $$
    \text{ReLU}(x) = \max(0, x)
    $$

### Derivadas de las Funciones de Activación

1. **Derivada de Sigmoid**: La derivada de la función Sigmoid se utiliza para calcular el gradiente durante el proceso de retropropagación en el entrenamiento de redes neuronales.

    $$
    \text{sigmoid}'(x) = \text{sigmoid}(x) \cdot (1 - \text{sigmoid}(x))
    $$

2. **Derivada de Tanh**: La derivada de la función Tanh también se utiliza en el cálculo del gradiente durante la retropropagación.

    $$
    \text{tanh}'(x) = 1 - \text{tanh}(x)^2
    $$

3. **Derivada de ReLU**: La derivada de la función ReLU es simple y eficiente de calcular.

    $$
    \text{ReLU}'(x) = 
    \begin{cases} 
    0 & \text{si } x \leq 0 \\
    1 & \text{si } x > 0 
    \end{cases}
    $$

A continuación, se muestran las gráficas de estas funciones de activación y sus derivadas.


```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

# Ejemplo de uso
x = np.linspace(-10, 10, 100)
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.plot(x, sigmoid(x))
plt.title('Sigmoid')

plt.subplot(1, 3, 2)
plt.plot(x, tanh(x))
plt.title('Tanh')

plt.subplot(1, 3, 3)
plt.plot(x, relu(x))
plt.title('ReLU')

plt.show()
def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.plot(x, sigmoid_derivative(x))
plt.title('Sigmoid Derivative')

plt.subplot(1, 3, 2)
plt.plot(x, tanh_derivative(x))
plt.title('Tanh Derivative')

plt.subplot(1, 3, 3)
plt.plot(x, relu_derivative(x))
plt.title('ReLU Derivative')

plt.show()
```
